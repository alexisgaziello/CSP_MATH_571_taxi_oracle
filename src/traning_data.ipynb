```python
df.drop(columns=['trip_start_timestamp'],inplace=True)
```

```python
cols = df.columns

categorical_cols = ['pickup_community_area','day','sky_level','year',
            'weekofyear','Quarter','Hour','Month','Day Name','sky_level','daytype']

numerical_cols =['Trips', 'temperature', 'relative_humidity','wind_direction','wind_speed','precipitation','Fare Last Month',
            'Tips Last Month','Trips Last Hour','Trips Last Week (Same Hour)','Trips 2 Weeks Ago (Same Hour)']

df[categorical_cols] = df[categorical_cols].apply(pd.Categorical) 

df[categorical_cols]
```

```python
df.info()
```

```python
from sklearn import preprocessing

x = df[['Fare Last Month','Trips Last Week (Same Hour)','Trips Last Hour','Tips Last Month',"Trips 2 Weeks Ago (Same Hour)",'temperature','relative_humidity','wind_direction','wind_speed','precipitation']].values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
dff = pd.DataFrame(x_scaled,columns=['Fare Last Month','Trips Last Week (Same Hour)','Trips Last Hour','Tips Last Month',"Trips 2 Weeks Ago (Same Hour)",'temperature','relative_humidity','wind_direction','wind_speed','precipitation'])
```

```python
dff
```

```python
df.drop(columns=['Fare Last Month','Trips Last Week (Same Hour)','Trips Last Hour','Tips Last Month',"Trips 2 Weeks Ago (Same Hour)",'temperature','relative_humidity','wind_direction','wind_speed','precipitation'],inplace=True)
```

```python
df.info()
```

```python
['Fare Last Month','Trips Last Week (Same Hour)','Trips Last Hour','Tips Last Month',"Trips 2 Weeks Ago (Same Hour)",'temperature','relative_humidity','wind_direction','wind_speed','precipitation']
```

```python
df = pd.concat([df, dff], axis=1, sort=False)
```

```python
df
```

```python
df=pd.get_dummies(df, columns=['sky_level','pickup_community_area','day','weekofyear','Quarter','Hour','Month','Day Name','daytype','year'],drop_first=True)

float64=['Fare Last Month','Trips Last Week (Same Hour)','Trips Last Hour','Tips Last Month',"Trips 2 Weeks Ago (Same Hour)",'temperature','relative_humidity','wind_direction','wind_speed','precipitation']

df[float64]=df[float64].astype('float32')

df.info()
```

```python
df.fillna(df.median(),inplace=True)
```
